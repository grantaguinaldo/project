### Energy Supply, Refinery

Notes from CA-TIMES modeling details.
- Refinery creep, or the amount that the existing stock of California's refineries is allowed to expand over time is `0.45%/year`.
- Refinery efficiency is kept constant at `89%` for all years in the model.
- CA-TIMES sets an upper limit on the amount of total refinery output that can come from a specific fuel. (see `Table 3.1` in the CA-TIMES modeling details).
