```python
#Elec.
tech_ = ['Electricity']

tech_sub = ['None']

source = ['California Import', 'California Domestic']

feedstocks = ['Coal', 'Oil', 'Natural Gas', 'Uranium', 
              'Wind', 'Solar', 'Hydro', 'Geothermal', 
              'Tidal', 'Forest Residues', 'MSW, Paper', 'MSW, Wood', 
              'MSW, Yard', 'Orchard and Vineyard Waste', 'Pulpwood', 
             'Ag. Residues (Stovers/Straws)', 'Energy Crops']

use = ['California Export', 'California Domestic']

products = ['Electricity', 'High Temp Heat', 'CO2 for Seq.']

product_del = ['None']

tech = ['Thermal Power Plants', 'CHP', 
        'Polygen Plants', 'Renewable Plants']

mix = []
for a in tech_:
    for b in tech_sub:
        for c in source:
            for d in tech:
                for e in use:
                    for f in feedstocks:
                        for g in products:
                            for h in product_del:
                                entry = {'Technology-Sector': a,
                                         'Technology-Sub': b,
                                         'Feedstock Source': c,
                                         'Technology': d,
                                         'Product Use': e,
                                         'Feedstock': f, 
                                         'Products': g,
                                         'Product Delivery Method': h}
                                mix.append(entry)
df_elec = pd.DataFrame(mix)
```
