```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect('US_9R_8D.sqlite')
cur = con.cursor()   
con.text_factory = str 
```

```python
query = "SELECT A.regions, A.tech, A.vintage, B.sector, B.tech_desc \
         FROM ExistingCapacity as A\
         INNER JOIN technologies as B \
         ON A.tech = B.tech \
         WHERE A.regions == 'CA'"

df = pd.read_sql(query, con)
df.shape
```

```python
query = "SELECT *\
         FROM technologies"

df = pd.read_sql(query, con)
df.shape
```

```python
query = "SELECT * \
         FROM technologies \
         WHERE sector LIKE '%ind%'"
df_ind_tech = pd.read_sql(query, con)
df_ind_tech.head() #technologies are inside of OEO.
```

```python
ind_tech = list(set(df_ind_tech['tech_desc'].tolist()))
ind_tech  #list of technologies are inside of OEO.
```

```python
query = "SELECT A.regions, A.tech, A.vintage, \
                A.exist_cap, B.tech_desc, B.sector \
         FROM ExistingCapacity as A \
         INNER JOIN technologies as B \
         ON A.tech = B.tech \
         WHERE A.regions == 'CA'"

df = pd.read_sql(query, con)
list_of_sectors_w_existing_cap = list(set(df['sector'].tolist()))
list_of_sectors_w_existing_cap # the industrial sector does not have existing capacities in OEO.
```
